```python
import numpy as np
# Use turtle when it comes to the actual drawing part

gridSize = 3
testGrid = [[0,1,2,3,4,5]
               ,['a','b','c','d','e','f']]



# Defining shapes
shp1 = np.array([[1,1],[1,0]])
shp2 = np.array([[1,0,0],[1,1,1]])
shp3 = np.array([[1]])
```

```python
startGrid = np.zeros((gridSize,gridSize),int)
solutionGrid = np.array(startGrid)

# Reshape arrays to size of grid
def reshape(square):
    n, m = square.shape
    bigsquare = np.zeros((gridSize, gridSize), square.dtype)
    bigsquare[0:n, 0:m] = square
    return(bigsquare)

# Try 4 configurations
def rotate(shp):
    return np.rot90(shp)

# Reflect and try other 4 configurations
def reflect(shp):
    return np.fliplr(shp)

# Function to shift values
def move(shp, dx,dy):
    n,m=shp.shape
    bigshp=np.zeros((3*n,3*m),shp.dtype) #Hmmm do these ones 
    bigshp[n:2*n,m:2*m]=shp
    x=n-dx
    y=m-dy
    return bigshp[x:x+n,y:y+m]

def valid(grid):
    if np.max(grid) > 1:
        return 1
    else:
        return 0

# If successful, lock in position else move to next grid location and repeat 8 configurations
i = 1
def testPosition(shp):
    global i
    while i == 1:
        global solutionGrid
        reshp = reshape(shp) # Reshape block array to size of whole array
        print("Printing reshaped shape")
        print(reshp)
        trialGrid = solutionGrid + reshp
        print("First trialGrid")
        print(trialGrid)
        test = valid(trialGrid)
        if test == 1:   # While the shapes overlap onto themselves 
            r = 0
            while (r < 4 and test == 1):
                i = 1
                reshp = rotate(reshp)
                print("Print reshp")
                print(reshp)
                r = r + 1
                print(r)
                trialGrid = solutionGrid + reshp
                print("Print trialGrid")
                print(trialGrid)
                test = valid(trialGrid)
                if test == 0: i = 0
                print("Rotating while valid value")
                print(test)
            reshp = move(reshp,1,0)
        else:
            i = 0
            solutionGrid = np.array(trialGrid)
            print("Print solutionGrid")
            print(solutionGrid)
    i = 1        
        


shps = [shp1, shp2, shp3]
        
for x in shps:
    testPosition(x)
    print(solutionGrid)
```
